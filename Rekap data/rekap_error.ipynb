```python
df00 = pd.read_excel('Error 10. Isian Subkelompok makanan jadi tidak sesuai.xls')
df01 = pd.read_excel('Error 11. Inkonsistensi Isian Bukan Makanan.xls')
df02 = pd.read_excel('Error 12. RH bukan makanan tidak sesuai.xls')
df03 = pd.read_excel('Error 13. Isian Subkelompok bukan makanan tidak sesuai.xls')
df04 = pd.read_excel('Error 14. Inkonsistensi Isian Out Of Pocket.xls')

df05 = pd.read_excel('Error 15. Pengeluaran Bukan Makanan Tidak Wajar.xls')
df06 = pd.read_excel('Error 17. Error jumlah komoditi makanan.xls')
df07 = pd.read_excel('Error 18. Error jumlah komoditi bukan makanan.xls')
df08 = pd.read_excel('Error 19. Error total komoditi.xls')
df09 = pd.read_excel('Error 2. Inkonsistensi Isian Bahan Makanan.xls')

df10 = pd.read_excel('Error 20. Error Under dan Over Kalori.xls')
df11 = pd.read_excel('Error 21. Duplikasi Pengeluaran Makanan.xls')
df12 = pd.read_excel('Error 22. Duplikasi Pengeluaran Bukan Makanan.xls')
df13 = pd.read_excel('Error 3. Konsumsi bahan makanan melebihi batas maksimum konsumsi.xls')
df14 = pd.read_excel('Error 3_1. Konsumsi bahan makanan melebihi batas maksimum konsumsi.xls')

df15 = pd.read_excel('Error 4. RH bahan makanan tidak sesuai.xls')
df16 = pd.read_excel('Error 5. Isian Subkelompok bahan makanan tidak sesuai.xls')
df17 = pd.read_excel('Error 6. ID ruta dengan jml art tidak sama dengan rekap R301.xls')
df18 = pd.read_excel('Error 7. Inkonsistensi Isian makanan jadi.xls')
df19 = pd.read_excel('Error 9. RH makanan jadi tidak sesuai.xls')
```

```python
# merge
df_merged = pd.concat([df00, df01, df02, df03, df04, df05, df06, df07, df08, df09,
                       df10, df11, df12, df13, df14, df15, df16, df17, df18, df19], axis=0)
```

```python
# sum by group
df_count = df_merged.groupby(['kab', 'Keterangan'])['kab'].count()
print(df_count)
```

```python
# convert to string
df_count = df_count.to_string()
```

```python
# write to txt in case of truncation
with open("output0.txt", "w") as f_out:
    f_out.write((df_count))
```
