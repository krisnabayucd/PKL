```python
import pandas as pd
```

```python
# read data
df = pd.read_excel('34.xlsx', header=None)
```

```python
df.head()
```

```python
# column name
df.columns = ['temuan']
```

```python
df.head()
```

```python
# extract last numbers to separate column
df['jumlah'] =  df['temuan'].str.extract('(\d+)$')
```

```python
df.head()
```

```python
import re
# define regex
regex_pat = re.compile(r'(\d+)$', flags=re.IGNORECASE)
```

```python
# remove last numbers from string
df['temuan'] = pd.Series(df['temuan']).str.replace(regex_pat, ' ', regex=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# extract df to xlsx
df.to_excel('34.xlsx')
```
